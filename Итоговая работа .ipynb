```python
import pandas as pd
import string
import re
import nltk
import pymorphy3

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer

import numpy as np
np.set_printoptions(threshold=np.inf)
```

### Чистка текста

```python
# Функция для удаления знаков препинания
def remove_punctuation(text): 
    return "".join([ch if ch not in string.punctuation else ' ' for ch in text])

# Функция для удаления чисел
def remove_numbers(text): 
    return ''.join([i if not i.isdigit() else ' ' for i in text])

# Функция для удаление множества пробелов
def remove_multiple_spaces(text): 
    return re.sub(r'\s+', ' ', text, flags=re.I)

# Функция для удаления спец. символов
st = '❯\xa0—«»'
def remove_othersymbol(text):
    return ''.join([ch if ch not in st else ' ' for ch in text])
```

Были созданы специальные методы, которые нужны для чистки текста.

```python
# Создание отдельного столбца для хранения чистого текста
df['clean_text'] = [remove_multiple_spaces(remove_numbers(remove_othersymbol(remove_punctuation(text.lower())))) for text in df['text']]
```

Был создат отдельный столбец clean_text для хранения очищенного текста. В нем мы удалили знаки препинания, все числа, лишние пробелы и специальные символы, а также, весь текст был переведен в нижний регистр. Для наглядности выведем неочищенный текст и чистый.

Теперь необхоимо удалить все стоп слова

```python
# Установка стоп-слов
import nltk
nltk.download('stopwords')
```

```python
# Сохраням русские стоп-слова
russian_stopwords = stopwords.words("russian") 
russian_stopwords
```

В данном списке хранятся все стоп-слова. Но в нем есть слова, которые удалять нельзя, такие как "не", "нет", "ни", "нельзя". Эти слова обозначают отрицание. Если их удалить то большинство твитов поменяют свою тональность.

```python
# Сохраняем в отдельный список слова, которые необходимо оставить
words_to_keep = ['не', 'нет', 'ни', 'нельзя']
# Оставляем в списке только те слова, которые не входят в words_to_keep
russian_stopwords = [word for word in russian_stopwords if word not in words_to_keep] 
```

Также в список стоп-слов нужно добавить технический мусор Twitter.

```python
# Создаем список с мусорными словами
twitter_garbage = ['http', 'https', 'co', 't', 'www', 'ru', 'rt', 'рт', 'это', 'че', 'чё', 'типа']

# Склеиваем два списка вместе
russian_stopwords = russian_stopwords + twitter_garbage
```

```python
# Функция для токенизации и удаления стоп-слов
def tokenize(text):
    t = word_tokenize(text)
    tokens = [token for token in t if token not in russian_stopwords]
    text = " ".join(tokens)
    return text
```

```python
# Выполняем токенизацию и удаляем стоп-слова в датасете
df['tokenize_text'] = [tokenize(text) for text in df['clean_text']]
```

```python
df.sample(5)
```

```python
# Для наглядности выведем первый твит
df['tokenize_text'][0]
```

Теперь выполним Лимитизацию для датасета, то есть привидем слова в начальную форму.

```python
morph = pymorphy3.MorphAnalyzer(lang='ru')
```

```python
%%time
lemm_texts_list = []
for text in (df['tokenize_text']):
    try:
        tokens = word_tokenize(text)
        res = list()
        for word in tokens:
            p = morph.parse(word)[0]
            res.append(p.normal_form)
        text = " ".join(res)
        lemm_texts_list.append(text)
    except Exception as e:
        print(e)
    
df['text_lemm'] = lemm_texts_list
```

```python
df.sample(5)
```

```python
# Выведем первый твит
df['text_lemm'][0]
```

Лимитизация прошла успешно. Все слова были переведены в первоначальную форму

```python
# Загружаем необхожимые библиотеки
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Получение текстовой строки из списка слов
def str_corpus(corpus):
    str_corpus = ''
    for i in corpus:
        str_corpus += ' ' + i
    str_corpus = str_corpus.strip()
    return str_corpus

# Получение списка всех слов в корпусе
def get_corpus(data):
    corpus = []
    for phrase in data:
        for word in phrase.split():
            corpus.append(word)
    return corpus

# Получение облака слов
def get_wordCloud(corpus):
    wordCloud = WordCloud(background_color='white',
                              stopwords=russian_stopwords,
                              width=3000,
                              height=2500,
                              max_words=200,
                              random_state=42
                         ).generate(str_corpus(corpus))
    return wordCloud
```

## Тематическое моделирование

```python
# Загрузка библиотек
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# Функция для определения темы
def get_topics(texts, label_name, num_topics=3, num_words=10):
    print(f"\n===== ТЕМАТИЧЕСКОЕ МОДЕЛИРОВАНИЕ ДЛЯ: {label_name} =====")
    
    vectorizer = CountVectorizer(max_features=1500, stop_words=russian_stopwords)
    X = vectorizer.fit_transform(texts)
    
    lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
    lda.fit(X)
    
    words = vectorizer.get_feature_names_out()
    for topic_idx, topic in enumerate(lda.components_):
        print(f"\n--> Тема №{topic_idx + 1}:")
        
        top_words_idx = topic.argsort()[:-num_words-1:-1]
        top_words = [words[i] for i in top_words_idx]
        print(", ".join(top_words))
```

```python
# 1. Выделяем тексты для позитива и негатива
pos_data_texts = df[df['tonality'] == 1]['text_lemm'].astype(str)
neg_data_texts = df[df['tonality'] == 0]['text_lemm'].astype(str)
```

```python
get_topics(pos_data_texts, "ПОЛОЖИТЕЛЬНЫЕ ТВИТЫ", num_topics=3)
get_topics(neg_data_texts, "ОТРИЦАТЕЛЬНЫЕ ТВИТЫ", num_topics=3)
```

- max_features=15000 - Берем топ-15000 самых важных слов и словосочетаний
- stop_words=russian_stopwords - Список стоп-слов
- ngram_range=(1, 2) - Берем одиночные слова и классные связки (типа "не_хочу", "очень_рад")

Судя по метрикам и ROC-кривой, модель обучилась очень даже хорошо.

```python
# Оценка качества обучения
y_pred_mul = mul.predict(x_test)
y_prob_mul = mul.predict_proba(x_test)
print('Метрики для Наивного Байеса')
print(classification_report(y_test,y_pred_mul))
```

Модель показала хорошую общую точность классификации (Accuracy = 73%). Алгоритм продемонстрировал сбалансированные показатели точности и полноты для обоих классов, при этом способность находить позитивные твиты оказалась чуть выше (recall = 0.75), чем негативные (recall = 0.71).

```python
# Выведем ROC-кривую
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

y_prob_mul = mul.predict_proba(x_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob_mul)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'k--')

plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve (Наивный Байес)')
plt.legend()
plt.show()
```

Модель также очень хорошо обучилась. 

По итогу, все три модели очень хорошо себя показали. У всех трех моделей почти равные значения метрик. Но все же Логистическая регрессия (LogisticRegression) показала результат чуть лучше, чем все остальные модельки.
